```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
header={
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.74 Safari/537.36"
}
request=requests.get("https://www.flipkart.com/oneplus-11r-5g-galactic-silver-256-gb/product-reviews/itmd8344a066fd54?pid=MOBGN3BQGU5BRTHZ&lid=LSTMOBGN3BQGU5BRTHZKJU8K7&marketplace=FLIPKART",headers=header)
request.text
```

```python
soup=BeautifulSoup(request.text)
soup
```

```python
stars=soup.find_all('div',{'class':'_3LWZlK _1BLPMq'})
stars_list=[]
for i in stars:
    stars_list.append(i.get_text())

len(stars_list)
```

```python
titles=soup.find_all('p',{'class':'_2-N8zT'})
titles_list=[]
for i in titles:
    titles_list.append(i.get_text())

len(titles_list)
```

```python
reviews=soup.find_all('div',{'class':'t-ZTKy'})
reviews_list=[]
for i in reviews:
    reviews_list.append(i.get_text())

len(reviews_list)
```

```python
likes=soup.find_all('span',{'class':'_3c3Px5'})
likes_list=[]
dislikes_list=[]
ct=0
for i in likes:
    if ct%2==0:
        likes_list.append(i.get_text())
        ct=1
    else:
        dislikes_list.append(i.get_text())
        ct=0

print(len(likes_list))
print(len(dislikes_list))
```

```python
data=pd.DataFrame()
data["stars"]=stars_list
data["titles"]=titles_list
data["reviews"]=reviews_list
data["likes"]=likes_list
data["dislikes"]=dislikes_list
data
```

```python
data.to_csv("ps30.csv")
```
